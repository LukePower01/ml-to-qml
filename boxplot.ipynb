```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

scores = pd.read_csv('bootstrap_scores.csv')
```

```python
scores.head()
```

```python
sns.set_theme(style = 'whitegrid')
```

```python
plt.figure(figsize=(13, 8))
df_score = pd.DataFrame(scores)
df = df_score.melt(var_name='Model', value_name='Score')

sns.boxplot(hue='Model', y='Score', data=df, palette=['blue', 'red', 'grey', 'green'],gap = .2)
plt.ylim(bottom=0.5)

plt.title('Distributions of Bootstrapped Scores')
plt.xlabel('Model')
plt.ylabel('Score')
plt.savefig('Visuals/models/Model_accuracy_boxplot_comparison.png')
plt.show()
```
